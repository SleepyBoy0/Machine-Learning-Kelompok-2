```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
print(os.listdir("../testing"))
```

```python
# import movie data set and look at columns
product = pd.read_csv("../testing/product.csv")
product.columns
```

```python
# what we need is that movie id and title
product = product.loc[:,["productid","nama"]]
product.head(10)
```

```python
# import rating data and look at columsn
rating = pd.read_csv("../testing/total.csv")
rating.columns
```

```python
# what we need is that user id, movie id and rating
rating = rating.loc[:,["userid","productid","total"]]
rating.head(100)
```

```python
# then merge movie and rating data
data = pd.merge(product,rating)
```

```python
# now lets look at our data 
data.head(100)
```

```python
data.shape
```

```python
data = data.iloc[:10000,:]
```

```python
pivot_table = data.pivot_table(index = ["userid"],columns = ["nama"],values = "total")
pivot_table.head(1000)
```

```python
product_rated = pivot_table["Gatsby Pomade"]
similarity_with_other_product = pivot_table.corrwith(product_rated)
similarity_with_other_product = similarity_with_other_product.sort_values(ascending=False)
similarity_with_other_product.head(4)
```
